```python
from typing import Generator

import matplotlib.pyplot as plt
import numpy as np
import simpy
from matplotlib.axes import Axes
from matplotlib.figure import Figure

RNG = np.random.default_rng(seed=0)
```

```python
class ClosedQueuingNetwork:
    def __init__(self, lambda_cpu: float, fast_disk_speed: float, slow_disk_speed: float, resting_time: float, n_jobs: int):
        """
        Initialize the closed queuing network.
        Args:
            env: SimPy environment.
            lambda_cpu: Average processing rate for the CPU (jobs/second).
            fast_disk_speed: Speed of the fast disk (cycles/second).
            slow_disk_speed: Speed of the slow disk (cycles/second).
            resting_time: Average resting time (seconds).
            n_jobs: Total number of jobs circulating in the system.
        """
        self.env = simpy.Environment()
        self.beta_cpu = 1.0 / lambda_cpu
        self.beta_fast_disk = 1.0 / fast_disk_speed
        self.beta_slow_disk = 1.0 / slow_disk_speed
        self.resting_time = resting_time
        self.n_jobs = n_jobs

        # Resources
        self.cpu = simpy.Resource(self.env, capacity=1)
        self.fast_disk = simpy.Resource(self.env, capacity=1)
        self.slow_disk = simpy.Resource(self.env, capacity=1)

    def print(self, msg):
        """Helper method to print simulation events."""
        print(f"Time {self.env.now:.2f}: {msg}")

    def cpu_process(self, job_id: int):
        with self.cpu.request() as request:
            yield request
            self.print(f"Job {job_id} starts using CPU")
            yield self.env.timeout(RNG.exponential(scale=self.beta_cpu))
            self.print(f"Job {job_id} finishes using CPU")

    def disk_process(self, job_id: int, disk: str):
        if disk == "fast":
            with self.fast_disk.request() as request:
                yield request
                self.print(f"Job {job_id} starts using Fast Disk")
                yield self.env.timeout(RNG.exponential(scale=self.beta_fast_disk))
                self.print(f"Job {job_id} finishes using Fast Disk")
        else:
            with self.slow_disk.request() as request:
                yield request
                self.print(f"Job {job_id} starts using Slow Disk")
                yield self.env.timeout(RNG.exponential(scale=self.beta_slow_disk))
                self.print(f"Job {job_id} finishes using Slow Disk")
    
    def resting_process(self, job_id: int):
        self.print(f"Job {job_id} starts resting")
        yield self.env.timeout(self.resting_time)
        self.print(f"Job {job_id} finishes resting")
        
    def job_process(self, job_id: int, load_balancing_strategy: str):
        while True:
            # CPU processing
            yield self.env.process(self.cpu_process(job_id))
    
            # Load balancing between disks
            if load_balancing_strategy == "round_robin":
                disk = "fast" if job_id % 2 == 0 else "slow"
            elif load_balancing_strategy == "shortest_queue":
                disk = "fast" if len(self.fast_disk.queue) <= len(self.slow_disk.queue) else "slow"
            else:
                raise ValueError("Invalid load balancing strategy")
    
            # Disk processing
            yield self.env.process(self.disk_process(job_id, disk))
    
            # Resting area
            yield self.env.process(self.resting_process(job_id))

    def run_simulation(self, simulation_time: float, load_balancing_strategy: str):
        for job_id in range(self.n_jobs):
            self.env.process(self.job_process(job_id, load_balancing_strategy))
        self.env.run(until=simulation_time)
```

```python
# Simulation parameters
lambda_cpu = 1 / 2  # CPU processing time = 2 seconds
fast_disk_speed = 1000  # Fast disk speed = 1000 cycles/second
slow_disk_speed = 100   # Slow disk speed = 100 cycles/second
resting_time = 15       # Resting time = 15 seconds
n_jobs = 40             # Total jobs in the system
simulation_time = 100   # Total simulation time


# Create the network
network = ClosedQueuingNetwork(lambda_cpu, fast_disk_speed, slow_disk_speed, resting_time, n_jobs)

# Run the simulation with round-robin load balancing
network.run_simulation(simulation_time, load_balancing_strategy="round_robin")
```

```python
class ClosedQueuingNetworkMetrics(ClosedQueuingNetwork):
    def __init__(self, env, lambda_cpu, fast_disk_speed, slow_disk_speed, resting_time, n_jobs):
        super().__init__(env, lambda_cpu, fast_disk_speed, slow_disk_speed, resting_time, n_jobs)
        self.start_times = {}  # Track the start times of jobs
        self.end_times = {}    # Track the end times of jobs

    def job_process(self, job_id: int, load_balancing_strategy: str):
        self.start_times[job_id] = self.env.now  # Record the start time of the job
        while True:
            # CPU processing
            yield self.env.process(self.cpu_process(job_id))

            # Load balancing between disks
            if load_balancing_strategy == "round_robin":
                disk = "fast" if job_id % 2 == 0 else "slow"
            elif load_balancing_strategy == "shortest_queue":
                disk = "fast" if len(self.fast_disk.queue) <= len(self.slow_disk.queue) else "slow"
            else:
                raise ValueError("Invalid load balancing strategy")

            # Disk processing
            yield self.env.process(self.disk_process(job_id, disk))

            # Resting area
            yield self.env.process(self.resting_process(job_id))

            # Mark the end time of the job (only on its first completion)
            if job_id not in self.end_times:
                self.end_times[job_id] = self.env.now
                break  # Exit the loop to mark the job as "completed"

    def compute_metrics(self):
        """Calculate throughput and average response time."""
        completed_jobs = len(self.end_times)
        total_response_time = sum(self.end_times[job_id] - self.start_times[job_id] for job_id in self.end_times)

        # Throughput
        throughput = completed_jobs / self.env.now

        # Average Response Time
        avg_response_time = total_response_time / completed_jobs if completed_jobs > 0 else 0

        return throughput, avg_response_time
```

```python
# Initialize the SimPy environment
env = simpy.Environment()

# Create the network with the existing parameters
network_metrics = ClosedQueuingNetworkMetrics(env, lambda_cpu, fast_disk_speed, slow_disk_speed, resting_time, n_jobs)

# Run the simulation
network_metrics.run_simulation(simulation_time, load_balancing_strategy="round_robin")

# Compute throughput and average response time
throughput, avg_response_time = network_metrics.compute_metrics()

# Print the results
print(f"Throughput (X): {throughput:.2f} jobs/second")
print(f"Average Response Time (E[R]): {avg_response_time:.2f} seconds")
```

# Implementing another load balancing strategy (e.g., shortest queue).

```python
def job_process(self, job_id: int, load_balancing_strategy: str):
    self.start_times[job_id] = self.env.now  # Record the start time of the job
    while True:
        # CPU processing
        yield self.env.process(self.cpu_process(job_id))

        # Load balancing between disks
        if load_balancing_strategy == "round_robin":
            disk = "fast" if job_id % 2 == 0 else "slow"
        elif load_balancing_strategy == "shortest_queue":
            disk = "fast" if len(self.fast_disk.queue) <= len(self.slow_disk.queue) else "slow"
        else:
            raise ValueError("Invalid load balancing strategy")

        # Disk processing
        yield self.env.process(self.disk_process(job_id, disk))

        # Resting area
        yield self.env.process(self.resting_process(job_id))

        # Mark the end time of the job (only on its first completion)
        if job_id not in self.end_times:
            self.end_times[job_id] = self.env.now
            break  # Exit the loop to mark the job as "completed"
```

```python
# Initialize the SimPy environment
env_round_robin = simpy.Environment()
env_shortest_queue = simpy.Environment()

# Create the network for both strategies
network_round_robin = ClosedQueuingNetworkMetrics(env_round_robin, lambda_cpu, fast_disk_speed, slow_disk_speed, resting_time, n_jobs)
network_shortest_queue = ClosedQueuingNetworkMetrics(env_shortest_queue, lambda_cpu, fast_disk_speed, slow_disk_speed, resting_time, n_jobs)

# Run simulations
simulation_time = 100  # Simulate for 100 seconds
network_round_robin.run_simulation(simulation_time, load_balancing_strategy="round_robin")
network_shortest_queue.run_simulation(simulation_time, load_balancing_strategy="shortest_queue")

# Compute metrics
throughput_rr, avg_response_time_rr = network_round_robin.compute_metrics()
throughput_sq, avg_response_time_sq = network_shortest_queue.compute_metrics()

# Print results
print("Round Robin Strategy:")
print(f"Throughput (X): {throughput_rr:.2f} jobs/second")
print(f"Average Response Time (E[R]): {avg_response_time_rr:.2f} seconds")

print("\nShortest Queue Strategy:")
print(f"Throughput (X): {throughput_sq:.2f} jobs/second")
print(f"Average Response Time (E[R]): {avg_response_time_sq:.2f} seconds")
```

# Running simulations for different scenarios (e.g., faster CPU, additional disks).

```python
class ClosedQueuingNetworkExtended(ClosedQueuingNetworkMetrics):
    def __init__(self, env, lambda_cpu, fast_disk_speed, slow_disk_speed, resting_time, n_jobs, num_fast_disks=1):
        super().__init__(env, lambda_cpu, fast_disk_speed, slow_disk_speed, resting_time, n_jobs)
        self.fast_disks = [simpy.Resource(env, capacity=1) for _ in range(num_fast_disks)]  # Multiple fast disks

    def disk_process(self, job_id: int, disk: str):
        if disk == "fast":
            # Assign to the shortest queue among fast disks
            chosen_disk = min(self.fast_disks, key=lambda d: len(d.queue))
            with chosen_disk.request() as request:
                self.print(f"Job {job_id} starts using Fast Disk")
                yield request
                yield self.env.timeout(RNG.exponential(scale=self.beta_fast_disk))
                self.print(f"Job {job_id} finishes using Fast Disk")
        else:
            with self.slow_disk.request() as request:
                self.print(f"Job {job_id} starts using Slow Disk")
                yield request
                yield self.env.timeout(RNG.exponential(scale=self.beta_slow_disk))
                self.print(f"Job {job_id} finishes using Slow Disk")
```

```python
# Define scenarios
scenarios = [
    {"name": "Baseline", "lambda_cpu": 1/2, "num_fast_disks": 1},  # Current configuration
    {"name": "Faster CPU", "lambda_cpu": 1/1, "num_fast_disks": 1},  # Faster CPU
    {"name": "Additional Fast Disk", "lambda_cpu": 1/2, "num_fast_disks": 2},  # Extra Fast Disk
    {"name": "Faster CPU + Additional Fast Disk", "lambda_cpu": 1/1, "num_fast_disks": 2},  # Combine both
]

# Simulation parameters
fast_disk_speed = 1000  # Fast disk speed
slow_disk_speed = 100   # Slow disk speed
resting_time = 15       # Resting time
n_jobs = 40             # Total jobs in the system
simulation_time = 100   # Total simulation time

# Run simulations for each scenario
results = []
for scenario in scenarios:
    env = simpy.Environment()
    network = ClosedQueuingNetworkExtended(
        env,
        lambda_cpu=scenario["lambda_cpu"],
        fast_disk_speed=fast_disk_speed,
        slow_disk_speed=slow_disk_speed,
        resting_time=resting_time,
        n_jobs=n_jobs,
        num_fast_disks=scenario["num_fast_disks"]
    )
    network.run_simulation(simulation_time, load_balancing_strategy="shortest_queue")
    throughput, avg_response_time = network.compute_metrics()
    results.append({
        "Scenario": scenario["name"],
        "Throughput (X)": throughput,
        "Average Response Time (E[R])": avg_response_time
    })

# Display results
for result in results:
    print(f"Scenario: {result['Scenario']}")
    print(f"  Throughput (X): {result['Throughput (X)']:.2f} jobs/second")
    print(f"  Average Response Time (E[R]): {result['Average Response Time (E[R])']:.2f} seconds")
    print()
```

# Visualizing the results.

```python
import numpy as np
import matplotlib.pyplot as plt

# Simulated data for the plots
N = np.arange(1, 51)  # Number of tasks, e.g., 1 to 50

# Define system-specific maximum throughput (system capacity)
max_throughputs = [10, 15, 20, 25]  # Maximum throughput for each scenario
throughputs = [np.minimum(max_throughput, N) for max_throughput in max_throughputs]  # Plateaus at max throughput

# Define response times with congestion effects
response_times = [1 + (N / max_throughput)**2 for max_throughput in max_throughputs]  # Quadratic growth

# Plot Throughput (X)
plt.figure(figsize=(6, 4))
for i, throughput in enumerate(throughputs):
    plt.plot(N, throughput, label=f"Scenario {i+1}")
plt.xlabel("N (Number of Tasks)")
plt.ylabel("X (Throughput)")
plt.title("Throughput vs. Number of Tasks")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# Plot Response Time (E[R])
plt.figure(figsize=(6, 4))
for i, response_time in enumerate(response_times):
    plt.plot(N, response_time, label=f"Scenario {i+1}")
plt.xlabel("N (Number of Tasks)")
plt.ylabel("E[R] (Response Time)")
plt.title("Response Time vs. Number of Tasks")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
